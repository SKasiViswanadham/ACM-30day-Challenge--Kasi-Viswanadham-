```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
df = pd.read_csv("/content/drive/MyDrive/Iris.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.drop(columns = ['Id'])
df.head()
```

```python
# plot fig
sns.pairplot(df, hue = 'Species')
plt.show()
```

```python
cat_columns = ['Species']
df = pd.get_dummies(df, columns = cat_columns , drop_first=True)
df.head()
```

```python
X = df.drop(columns=['Species_Iris-versicolor', 'Species_Iris-virginica'])
```

```python
inertia = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.plot(k_range, inertia, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clustes')
plt.ylabel('Inertia')
plt.show()
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

kmeans_pca = KMeans(n_clusters=2, random_state=42)
clusters_pca = kmeans_pca.fit_predict(X_pca)

plt.figure(figsize=(8,5))
for cluster in range(2):
    plt.scatter(
        X_pca[clusters_pca == cluster, 0],
        X_pca[clusters_pca == cluster, 1],
        label = f'Cluster {cluster}'
    )
plt.title("K-Means After PCA")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
dendrogram = sch.dendrogram(sch.linkage(X_pca, method='ward'))
plt.title("Dendrogram")
plt.xlabel('samples')
plt.ylabel('Distance')
plt.show()
```

```python
hc = AgglomerativeClustering(n_clusters=2, metric='euclidean')
hc_labels = hc.fit_predict(X_pca)
```

```python
plt.figure(figsize=(8,5))
for cluster in range(2):
    plt.scatter(X_pca[hc_labels == cluster, 0],
                X_pca[hc_labels == cluster, 1],
                label = f'Cluster {cluster}'
               )
plt.title("Hierarchical Clustering")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend()
plt.grid(True)
plt.show()
```
